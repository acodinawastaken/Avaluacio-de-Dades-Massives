```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# data directory
data_dir = "../Dades-Notes/"
```

```python
df_sise = pd.read_csv(data_dir+"SisePrim.csv", low_memory=False)
df_sise = df_sise.replace(',', '.', regex=True)
df_sise.pop("PFRAN")
df_sise.pop("PFRAN_CO")
df_sise.pop("PFRAN_CE")
df_sise.pop("PFRAN_EE")
```

```python
df_sise.sort_values(by="AREA_TERRITORIAL", ascending=True)
```

```python
df_sise[["AREA_TERRITORIAL", "ANY"]].groupby('AREA_TERRITORIAL').count()
```

```python
df_sise2009 = df_sise[df_sise["ANY"]==2009]
df_sise2009[["NATURALESA","ANY"]].groupby('NATURALESA').count()
```

```python
def toFloat(x):
    try:
        x = float(x) # from string to float to int
    except:
        x = np.NaN
    return x
```

```python
df_sise["PCAT"] = df_sise["PCAT"].astype(float)
df_sise["PCAST"] = df_sise["PCAST"].astype(float)
```

```python
Arees=["Baix Llobregat","Barcelona Comarques","Catalunya Central","Consorci d'Educació de Barcelona","Girona","Lleida","Maresme - Vallès Oriental","Tarragona","Terres de l'Ebre","Vallès Occidental"]
Assignatures=["PCAT","PCAST","PANG","PMAT","PMED"]
for i in Arees:
    df_sise2009_Area= df_sise2009[df_sise2009["AREA_TERRITORIAL"]==i]
    for j in Assignatures:
        df_sise2009_Area[j] = df_sise2009_Area[j].astype(float)
        average_value = df_sise2009_Area[j].mean(skipna=True)
        print(f'The average value in {j} in {i} is {average_value}')
```
